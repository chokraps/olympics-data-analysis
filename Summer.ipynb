## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
goldMedals = df[df.Medal == 'Gold']
data = []
for sport in goldMedals['Sport'].unique():
    data.append([sport, len(goldMedals[goldMedals['Sport'] == sport])])
    
data = pd.DataFrame(data, columns = ['Sport', 'Medal'])
data = data.sort_values(by = 'Medal', ascending = False).head()
print(data)
data.plot( x = 'Sport', y = 'Medal', kind = 'bar')
```
